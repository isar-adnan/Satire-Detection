```python
import os
import sys

class PolarityFinder:

    def __init__(self):

        self.pos_words = self._load_words( "./pos.txt")
        self.neg_words= self._load_words( "./neg.txt")



    def _load_words(self , filename):

        all_words = []

        with open(filename) as f:

            for line in f.readlines():
                all_words.append(line.strip())

            return all_words

    def clear_text(self , text):

        ignore_list = ["," , "''" , "''" , "." ,  "\""]
        new_text = ""

        for char in text:
            if char in ignore_list:
                continue
            new_text += char

        return new_text.strip()


    def show_sentiment_of(self , filename):

        pos_count = 0
        neg_count = 0

        with open(filename) as f:
            lines = f.readlines()

            for line in lines:
                if len(line.strip()) == 0:
                    continue

                clean_line = self.clear_text(line)

                for word in clean_line.split(" "):
                    if word in self.pos_words:
                        pos_count += 1
                    elif word in self.neg_words:
                        neg_count += 1


            print("===============================================================")
            print("Positive Sentiment {} = {}".format(filename , pos_count)) 
            print("Negative Sentiment {} = {}".format(filename , neg_count))
            print("Sentiment Polarity {} = {}".format(filename , pos_count - neg_count))
            print("================================================================")



def main():

    

    dirpath = "test"

    files = os.listdir(dirpath)

    polarity_finder = PolarityFinder()

    for current_file in files:
        file_path = dirpath + os.sep + current_file
        polarity_finder.show_sentiment_of(file_path)




if __name__ == "__main__":
    main()
```
